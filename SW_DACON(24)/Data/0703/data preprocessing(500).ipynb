```python
import numpy as np
import librosa

sample_rate = 24000

for i in range(train_feature.shape[0]):
    x = librosa.load(train_path + train_list[i])[0]
    S = librosa.feature.melspectrogram(y=x, sr=sample_rate, n_mels=128)
    log_S = librosa.power_to_db(S, ref=np.max)
    mfcc = librosa.feature.mfcc(S=log_S, n_mfcc=20)
    mfcc_shape = mfcc.shape[1]

    # width 값을 mfcc_shape에 맞게 조정, 홀수이고 3 이상이어야 함
    if mfcc_shape < 3:
        width = 3  # mfcc_shape가 3보다 작으면 최소값 3으로 설정
    else:
        width = min(9, mfcc_shape - 1)  # mfcc_shape가 충분히 크면 9 이하로 조정
        if width % 2 == 0:
            width -= 1  # width가 짝수이면 홀수로 조정
    
    n_mfcc = librosa.feature.delta(mfcc, order=2, width=width)
    n_mfcc = n_mfcc.flatten()
    n_mfcc = n_mfcc[:500]
    
    for j in range(train_feature.shape[1]):
        if n_mfcc.shape[0] < 500:
            break
        train_feature.iloc[i, j] = n_mfcc[j]
```
