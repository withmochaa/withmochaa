```python
import numpy as np
import pandas as pd
import librosa

# Constants
duration = 2  # seconds
sample_rate = 16000  # Hz

# train_feature 데이터프레임의 크기에 맞게 수정
train_feature = pd.DataFrame(np.zeros((train_feature.shape[0], 500)))  # 예시로 train_feature 초기화

for i in range(train_feature.shape[0]):
    # Load audio file
    x, _ = librosa.load(train_path + train_list[i], sr=sample_rate, duration=duration)
    
    # Extract features
    S = librosa.feature.melspectrogram(y=x, sr=sample_rate, n_mels=128)
    log_S = librosa.power_to_db(S, ref=np.max)
    
    mfcc = librosa.feature.mfcc(S=log_S, n_mfcc=20)
    
    # Adjust width for delta calculations
    delta_width = min(3, mfcc.shape[1])
    
    delta_mfcc = librosa.feature.delta(mfcc, order=1, width=delta_width)
    delta2_mfcc = librosa.feature.delta(mfcc, order=2, width=delta_width)
    
    zcr = librosa.feature.zero_crossing_rate(y=x)
    spectral_centroid = librosa.feature.spectral_centroid(y=x, sr=sample_rate)
    spectral_rolloff = librosa.feature.spectral_rolloff(y=x, sr=sample_rate)
    chroma_stft = librosa.feature.chroma_stft(y=x, sr=sample_rate)
    
    features = np.concatenate((mfcc, delta_mfcc, delta2_mfcc, zcr, spectral_centroid, spectral_rolloff, chroma_stft), axis=0)
    
    # Flatten and truncate or pad the feature vector to length 300
    features = features.flatten()
    if features.shape[0] > 500:
        features = features[:500]
    elif features.shape[0] < 500:
        features = np.pad(features, (0, 500 - features.shape[0]), mode='constant')

    # Assign the features to the dataframe
    for j in range(train_feature.shape[1]):
        train_feature.iloc[i, j] = features[j]

# train_feature 데이터프레임을 이후에 모델 학습 등에 사용
```
