```python
import numpy as np
import pandas as pd
import librosa

# Constants
duration = 5  # seconds
sample_rate = 32000  # Hz
n_mels = 128
n_mfcc = 20

# 임시로 큰 값을 사용하여 모든 특성을 유지
train_feature = []

for i in range(len(train_list)):
    x, _ = librosa.load(train_path + train_list[i], sr=sample_rate, duration=duration)
    

    if len(x) < 512:
        n_fft = len(x)
    else:
        n_fft = 512
        
    if n_fft==1024:
        n_fft=512
    
    # Extract features
    S = librosa.feature.melspectrogram(y=x, sr=sample_rate, n_mels=n_mels, n_fft=n_fft)
    log_S = librosa.power_to_db(S, ref=np.max)
    
    mfcc = librosa.feature.mfcc(S=log_S, n_mfcc=n_mfcc, n_fft=n_fft)
    
    # Adjust width for delta calculation
    width = min(9, mfcc.shape[1])
    if width < 3:
        width = 3
    elif width % 2 == 0:
        width += 1
    
    delta_mfcc = librosa.feature.delta(mfcc, order=1, width=width)
    delta2_mfcc = librosa.feature.delta(mfcc, order=2, width=width)
    
    zcr = librosa.feature.zero_crossing_rate(y=x)
    spectral_centroid = librosa.feature.spectral_centroid(y=x, sr=sample_rate)
    spectral_bandwidth = librosa.feature.spectral_bandwidth(y=x, sr=sample_rate)
    spectral_contrast = librosa.feature.spectral_contrast(S=log_S, sr=sample_rate)
    spectral_rolloff = librosa.feature.spectral_rolloff(y=x, sr=sample_rate)
    chroma_stft = librosa.feature.chroma_stft(y=x, sr=sample_rate)
    tonnetz = librosa.feature.tonnetz(y=librosa.effects.harmonic(x), sr=sample_rate)
    
    features = np.concatenate((mfcc, delta_mfcc, delta2_mfcc, zcr, spectral_centroid, spectral_bandwidth, spectral_contrast, spectral_rolloff, chroma_stft, tonnetz), axis=0)
    
    features = features.flatten()
    train_feature.append(features)

```

```python
max_feature_length = max(len(f) for f in train_feature)
train_feature_padded = np.zeros((len(train_feature), max_feature_length))
for i, f in enumerate(train_feature):
    train_feature_padded[i, :len(f)] = f

train_feature_df = pd.DataFrame(train_feature_padded)
```
